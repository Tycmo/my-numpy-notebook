```python
import pandas as pd
```

# index values shape ndim size name dtype

```python
s=pd.Series([1,2,3,"454",56],index=["a","z","x","c","v"])
print(s)
```

```python
print(s.index)
print(s.values)
print(s.shape)
print(s.ndim)
print(s.size)
s.name="1314520"
print(s.name)
print(s.dtype)
```

# loc[]  iloc[]  at[] iat[]

```python
print(s.loc["a":"z"])#显示索引
print("\n")
print(s.iloc[1:4])   #隐式索引
print(s.at["a"])    #比loc低无法切片
print(s.iat[1])
```

# 布尔索引

```python
print(s)
s.iloc[3]=64
print(s[s>10])
print(s.head())  #返回前五行数据
print(s.tail())  #返回后五行
```
