```python
import numpy as np
```

```python
arr=np.random.randint(1,25,10)
print(arr)
```

```python
print(np.sum(arr))  #求和
```

```python
print(np.mean(arr))  #求平均值
```

```python
print(np.median([1,2,4]))   #中位数
print(np.median([1,2,2,4,5,9]))
```

```python
print(np.var([1,2,3,4]))  #方差
print(np.std([1,2,3,4]))  #标准差
```

```python
np.random.seed(10)
arr=np.random.randint(1,100,20)
print(arr)
print(np.max(arr),np.argmax(arr))
print(np.min(arr),np.argmin(arr))
```

```python
print(np.percentile(arr,20))   #分位数
```

```python
arr=np.array([1,2,3,4])
print(np.sum(arr))
print(np.cumsum(arr))  #累计和
print(np.cumprod(arr)) #累积积
```
