# 1.平方根

```python
import numpy as np
print(np.sqrt(81),"\n")
print(np.sqrt([1,4,9]),"\n")
print(np.sqrt(np.array([[1,4,9],[4,9,9]])),"\n")
```

# 2.指数  e^x

```python
print(np.exp(3))
print(np.exp(1))
```

# 3.自然对数  ln x

```python
print(np.log(2.71821))
print(np.log(1))
```

# 4.三角函数

```python
print(np.sin(1))
print(np.sin(np.pi/2))
```

# 5.绝对值

```python
arr=np.array([-1,-6,5,-2])
print(np.abs(arr))
```
