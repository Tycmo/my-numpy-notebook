# 1.温度 
求  平均 max min   超过30度的天数

```python
import numpy as np
np.random.seed(10)
arr=np.random.randint(19,36,24)
print(arr)
```

```python
print("平均气温")
print("%.3f" %np.mean(arr))
print("最高气温")
print("%d" %np.max(arr))
print("最低气温")
print("%d" %np.min(arr))
print("超过30度的天数")
print(len(arr[arr>30]))
print(np.sum(np.where(arr>30,1,0)))
print(np.cumsum(np.where(arr>30,1,0))[-1])
print(np.count_nonzero(arr>30))
```

# 2.成绩
计算平均分  中位数   标准差 
将成绩转化为百分制

```python
np.random.seed(25)
score=np.random.randint(80,100,10)
print(score)
print(np.sort(score))
```

```python
print("平均数: %.2f" %np.mean(score))
print("中位数：%d" %np.median(score))
print("标准差：%.2f" %np.std(score))
```

# 4.随机数的计算 
行 列  替换

```python
np.random.seed(25)
arr=np.random.randint(0,10,(3,4))
print(arr)
```

```python
print(np.max(arr,axis=0))  #列
print(np.max(arr,axis=1))  #行
print(np.min(arr,axis=1))
print(np.min(arr,axis=0))
```

```python
print(np.where(arr%2!=0,-1,arr))
arr[arr%2!=0]=-1
print(arr)
```

# 5.数组变形

```python
arr=np.arange(1,13)
print(arr)
```

```python
arr=np.reshape(arr,(3,4))
print(arr)
print("列数最大值",np.sum(arr,axis=0))
print("行数最小值",np.sum(arr,axis=1))
print("列平均值",np.mean(arr,axis=0))
print("行平均值",np.mean(arr,axis=1))
```

# 7.统计应用

最大索引  最小索引

```python
np.random.seed(25)
arr=np.random.randint(100,200,9)
print(arr)
```

```python
print(np.sum(arr))
print(np.mean(arr))
print(np.var(arr))
print(np.std(arr))
print(np.argmax(arr)+1)
print(np.argmin(arr)+1)
```

# 9.唯一值和排序
确定唯一值排序
确定出现次数

```python
np.random.seed(25)
arr=np.random.randint(1,5,8)
print(arr)
```

```python
arr1,arr2=np.unique(arr,return_counts=True)
print(arr1)
print(arr2)
d=[]
```

```python
for i in range(len(arr1)):
    d=d+[len(arr[arr==arr1[i]])]
print(d)
```

# 10.总和
## 销售 成本 
### 利润 平均值 标准差 利润最高天数

```python
销售=np.random.randint(20,28,8)
成本=np.random.randint(15,21,8)
print(销售)
print(成本)
```

```python
利润=销售-成本
print(利润)
print("平均值：",np.mean(利润))
print("标准差:",np.std(利润))
print("利润最高天数:",np.argmax(利润))
```
