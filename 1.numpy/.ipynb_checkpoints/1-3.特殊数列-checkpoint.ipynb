### 等差数列   （start end step）   包前不包后

```python
import numpy as np
arr=np.arange(1,10,2)
print(arr)
arr=np.arange(2,20,1)
print(arr)
```

### 等间隔数列 （start  end size ) 包前包后

```python
arr =np.linspace(1,20,5)
print(arr)
arr =np.linspace(0,4,3,dtype=int)
print(arr)
arr=np.arange(0,101,20)
print(arr)
arr=np.linspace(0,100,6,dtype=int)
print(arr)
```

### 对数间隔数列

```python
arr=np.logspace(2,20,22,base=6,dtype=int)
print(arr)
```

```python
arr=np.logspace(0,10,10,dtype=int)
print(arr)
```
