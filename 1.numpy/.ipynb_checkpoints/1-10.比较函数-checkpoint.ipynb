```python
import numpy as np
```

```python
arr=[1,2,3,4,5,6,7,8,9]
print(np.greater(arr,5))   #判断断是否大于
print(np.less(arr,5))      #判断是否小于
print(np.equal(arr,5))     #判断是否等于
```

```python
print(np.logical_and([1,0],[1,1]))   #逻辑和
print(np.logical_or([1,0],[1,0]))    #逻辑or
print(np.logical_not([1,0]))         #逻辑not
```

```python
print(np.any([1,0,0,0]))   #判断是否至少有一个true
print(np.all([1,0,0,0]))   #判断是否all为true
```

```python
arr=np.random.randint(50,100,24)
print(arr)
print(np.where(arr<60,"不及格",
               np.where(arr<85,"良好","优秀")))
```

```python
print(np.select([arr<60,(arr>=60) & (arr<=84),arr>=85],["不及格","良好","优秀"],default="未知"))
```
