## 单位矩阵

```python
import numpy as np
```

```python
arr=np.eye(6,dtype=int)
print(arr)
print("\n")
arr=np.eye(4,5,dtype=int)
print(arr)
```

## 对角矩阵

```python
arr=np.diag([5,3,4,6])
print(arr)
```

## 随机矩阵

### 0~1 均匀 浮点 分布

```python
arr=np.random.rand(2,3)
print(arr)
```

### 指定范围区间  均匀 浮点 分布

```python
arr=np.random.uniform(3,6,(5,5))
print(arr)
```

```python
arr=np.random.randint(3,6,(5,5))
print(arr)
```

### 随机范围区间（-3，3） 正态分布

```python
arr=np.random.randn(6,4)
print(arr)
```

### 随机种子 确立种子后 固定随机数

```python
np.random.seed(20)   
arr=np.random.randint(1,10,(5,5))
print(arr)
print("\n")
arr=np.random.randint(5,79,(6,6))
print(arr)
```
