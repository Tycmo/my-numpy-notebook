```python
import numpy as np
```

```python
np.random.seed(100)    
arr=np.random.randint(1,50,20)
print(arr)
print(np.sort(arr))    #从小到大排序，不破坏原数组
print(arr)
print(np.argsort(arr))  #给出各元素的排名
arr.sort()
print(arr)               #从小到大排序，破坏原数组
```

```python
print(np.unique(arr))  #去重
```

```python
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[4,5,6],[7,8,9]])
print(arr1+arr2)
print(np.concatenate((arr1,arr2)))  #拼接
```

```python
print(np.split(arr,5)) #分割
print(np.split(arr,[2,8,10,14,19]))
```

```python
print(np.reshape(arr,[5,4]))
```
