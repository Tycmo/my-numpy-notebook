# 加减乘除

```python
import numpy as np
```

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
print(a*100)
print(a+100)
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

# 广播
# 同一维度或其中一个是一

```python
a=np.array([1,2,3])  #1*3
b=np.array([[4],[5],[6]])  #3*1
print(a+b)
print(a-b)
print(a*b)
print(a@b)      #线性代数乘法运算
```
