# (2) copy之前的

# （3）已有的结构

## 全零形

```python
arr00=np.full((5,6),8)
arr=np.zeros((2,3))
arr01=np.zeros_like(arr00)
print(arr,"\n")
print(arr01)
```

## 全一形

```python
arr1=np.ones((2,3))
arr10=np.ones_like(arr00)
print(arr1,"\n")
print(arr10)
```

## 随机

```python
arr2=np.empty((6,3))
arr20=np.empty_like(arr00)
print(arr2,"\n")
print(arr20)
```

### 自定义

```python
arr3=np.full((5,3),6)
arr30=np.full_like(arr00,100)
print(arr3,"\n")
print(arr30)
```
