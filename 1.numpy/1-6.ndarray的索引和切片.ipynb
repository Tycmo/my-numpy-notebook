```python
import numpy as np
```

# 一维数组

```python
arr =np.random.randint(1,100,20)
print(arr)
```

```python
print(arr[:])
print(arr[2:6])  #左闭右开
print(arr[slice(2,6)])
print(arr[(arr>90) & (arr<100)])
for i in range(20):
    print(arr[i])
```

## 二维数组

```python
arr=np.random.randint(1,100,(5,6))
print(arr)
```

```python
print(arr[1,2])
print(arr[1,:])#行数据
print(arr[2])
print(arr[:,2]) #列数据
print(arr[arr>50])
```
