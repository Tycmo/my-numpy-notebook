# 布尔 bool
# 整型 int uint  8 16 32 64 
# 浮点 float 16 32 64 
# 复数 complex  64 128

```python
import numpy as np
arr=np.array([2,0,1,3],dtype=np.bool)
print(arr)
```

```python
arr=np.array([6,5,8,9000],dtype=np.int16)
print(arr)
```

```python
arr=np.array([5.652,8.546,6.5544545,9.455454],dtype=np.float16)
print(arr)
```

```python
arr=np.array([5.445645484844,6.544554444455,9.5545445555],dtype=np.float32)
print(arr)
```

```python
arr=np.array([5.15546555555,8.55555555888881111122,6.555555587777777,9],dtype=np.float64)
print(arr)
```

```python
arr=np.array([5,8,6,9],dtype=np.complex64)
print(arr)
```

```python
arr=np.array([5,8,6,9],dtype=np.complex128)
print(arr)
```
