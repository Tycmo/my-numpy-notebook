```python
import numpy as np
```

# 1. 将一维数组保存为二进制`.npy`文件，再读取回来

```python
arr=np.array([1,2,3])
np.save("array.npy",arr)
loaded=np.load("array.npy")
print(loaded)
```

# 2. 将多个数组保存为压缩的`.npz`文件，并读取。

```python
np.random.seed(520)
arr1=np.random.randint(0,100,(10,21))
arr2=np.random.randint(0,100,(10,21))
arr3=np.random.randint(0,100,(10,21))
arr4=np.random.randint(0,100,(10,21))
arr5=np.random.randint(0,100,(10,21))
np.savez("arrays.npz",arr1,arr2,arr3,arr4,arr5)
loads=np.load("arrays.npz")
print(loads["arr_0"])
```

# 3. 将二维数组保存为文本文件（如CSV），再用`loadtxt`读取。

```python
np.random.randint(520)
arr=np.random.randint(0,10,(5,2))
print(arr)
np.savetxt("arr.csv",arr,delimiter=",")
loads=np.loadtxt("arr.csv",delimiter=",")
print(loads)
```

# 4. 使用`genfromtxt`读取带缺失值的文本文件。 

```python
with open("txt.csv","w") as f:
    f.write("1,2,3\nN,4,5\n7,8,9")
data=np.genfromtxt("txt.csv",delimiter=",",missing_values="N",filling_values=0)
print(data)
              
```

# 5. 将数组保存为TXT格式，并指定分隔符为逗号。

```python
arr=np.array([1,2,3])
np.savetxt("a.txt",arr,delimiter=",")
lopads=np.loadtxt("a.txt",delimiter=",")
print(lopads)
```
