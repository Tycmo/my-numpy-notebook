```python
import numpy as np 
```

# 1. 对数组每个元素求平方根。

```python
arr=np.array([1,36,49,81])
print(np.sqrt(arr))
```

# 2. 对数组每个元素求自然对数。

```python
print(np.log(arr))
```

# 3. 计算两个数组的对应元素之和、差、积、商。

```python
a=np.arange(1,11)
b=np.arange(1,11)
print(a)
print(b)
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

# 4. 计算数组的三角函数（sin, cos）并比较。

```python
arr=np.linspace(0,np.pi/2,5)
print(arr)
print(np.sin(arr))
print(np.cos(arr))
```

# 5. 使用`exp`计算e的幂。

```python
print(np.exp(5))
```

# 6. 对数组四舍五入到小数点后2位

```python
np.random.seed(520)
arr=np.random.uniform(0,12,12)
print(arr)
print(np.round(arr,2))
```

# 7. 使用`modf`将数组拆分成整数和小数部分

```python
print(np.modf(arr))
```

# 8. 实现数组的逐元素比较（大于、小于、等于），返回布尔数组。

```python
np.random.seed(520)
a=np.random.randint(-150,150,10)
b=np.random.randint(-150,150,10)
print(a)
print(b)
print(">",a>b)
print("<",a<b)
print("=",a==b)
```

# 9. 使用`where`实现条件赋值：将数组中大于0.5的元素替换为1，否则为0。

```python
print(np.where(a>0.5,1,0))
```

# 10. 计算数组的绝对值、符号函数（`sign`）。

```python
print(np.abs(a))
print(np.sign(a))
```
