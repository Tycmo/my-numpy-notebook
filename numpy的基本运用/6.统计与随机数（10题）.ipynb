```python
import numpy as np
```

# 1. 计算数组的均值、中位数、标准差

```python
arr=np.arange(12)
print(arr)
print("均值:",np.mean(arr))
print("中位数",np.median(arr))
print("标准差",np.std(arr))
```

# 2. 计算数组的方差、最小值、最大值

```python
print("方差",np.var(arr))
print("最小值",np.min(arr))
print("最大值",np.max(arr))
```

# 3. 对矩阵按列求和、按行求均值。

```python
np.random.seed(520)
arr=np.random.randint(1,100,(4,5))
print(arr)
for i in range(4):
    print("第%d行和"%(i+1),np.sum(arr[i,:]))
for i in range(5):
    print("第%d列平均值"%(i+1),np.mean(arr[:,i]))
print(np.sum(arr,axis=0))
print(np.mean(arr,axis=1))
```

# 4. 计算数组的累积和（`cumsum`）和累积积（`cumprod`）。

```python
print("累计和",np.cumsum(arr))
print("累积积",np.cumprod(arr))
```

# 5. 生成1000个服从标准正态分布的随机数，并计算均值和方差

```python
np.random.seed(520)
arr=np.random.randn(1000)
print(arr)
print(np.mean(arr))
print(np.var(arr))
```

# 6. 生成1000个服从均匀分布U(0,1)的随机数，绘制直方图（用matplotlib）。

```python
import matplotlib.pyplot as plt

samples = np.random.random(1000)
plt.hist(samples, bins=30, edgecolor='black')
plt.title('Uniform Distribution [0,1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# 7. 从给定数组`[1,2,3,4,5]`中随机抽取10个样本（可重复）。

```python
arr=np.array([1,2,3,4,5])
print(np.random.choice(arr,size=10,replace=True))
```

# 8. 打乱一个数组的顺序（`shuffle`或`permutation`）。

```python
print(np.random.permutation(arr))
np.random.shuffle(arr)
print(arr)   #就地修改
```

# 10. 生成一个5×5的随机整数矩阵，并找到每行的最大值及其位置。

```python
np.random.seed(520)
arr=np.random.randint(0,520,(5,5))
print(arr)
print(np.max(arr,axis=1,))
print(np.argmax(arr,axis=1))
```
