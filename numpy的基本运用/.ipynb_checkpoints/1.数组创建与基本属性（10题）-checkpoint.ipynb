```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. 创建一个长度为10的一维全零数组。

```python
arr=np.zeros(10,dtype=int)  #1
print(arr)
arr=np.empty(10,dtype=int) #2
arr.fill(0)
print(arr)
plt.plot(arr)
```

# 2. 创建一个3×4的全1矩阵，数据类型为int32。


```python
arr=np.ones((3,4),dtype=np.int32)  #1
print(arr)
arr=np.empty((3,4),dtype=np.int32)  #2
arr.fill(1)
print(arr)
arr=np.full((3,4),1,dtype=np.int32) #3
print(arr)
```

# 3. 使用`arange`生成从5到50、步长为5的数组。

```python
arr=np.arange(5,50,5)
print(arr)
```

# 4. 使用`linspace`生成0到π之间20个点的数组。

```python
arr=np.linspace(0,np.pi,20)
print(arr)
```

# 5. 创建一个3×3的单位矩阵。

```python
arr=np.eye(3,dtype=np.int64)  #1
print(arr)
arr=np.identity(3,dtype=np.int64)  #2.identity 只用于方阵
print(arr)
arr=np.diag([1,1,1])     #3
print(arr)
```

# 6. 创建一个对角线上元素为[1,2,3,4]的4×4矩阵。

```python
arr=np.diag([1,2,3,4])  #1
print(arr)
arr=np.zeros((4,4))
np.fill_diagonal(arr,[1,2,3,4])  #2
print(arr)
```

# 7. 创建一个随机整数数组，形状为(4,5)，范围在0-100之间。

```python
np.random.seed(520)
arr=np.random.randint(0,100,(4,5))
print(arr)
plt.imshow(arr)
```

# 8. 创建一个3×3×2的三维随机数组（均匀分布）。

```python
arr=np.random.random((3,3,2))
print(arr)
```
