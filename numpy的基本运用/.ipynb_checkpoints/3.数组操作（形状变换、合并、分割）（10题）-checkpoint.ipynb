```python
import numpy as np
```

# 1. 将一维数组(12,)重塑为3×4矩阵。

```python
np.random.seed(520)
arr=np.random.randint(-12,12,12)
print(arr)
arr2=arr.reshape(3,4)
print(arr2)
```

# 2. 将3×4矩阵展平为一维数组。

```python
arr2=np.random.rand(3,4)
print(arr2)
print(arr2.reshape(-1))  #1
print(arr2.flatten())    #2
print(arr2.ravel())      #3
```

# 3. 创建两个2×3矩阵，垂直堆叠成4×3。

```python
A=np.random.randint(1,100,(2,3))
B=np.random.randint(1,100,(2,3))
print(A)
print(B)
print(np.vstack((A,B)))  #1
print(np.concatenate((A,B),axis=0))  #2
```

# 4. 创建两个3×2矩阵，水平堆叠成3×4

```python
arr1=np.random.randint(0,100,(3,2))
arr2=np.random.randint(0,100,(3,2))
print(arr1)
print(arr2)
print(np.hstack((arr1,arr2))) #1
arr=np.hstack((arr1,arr2))
print(np.concatenate((arr1,arr2),axis=1)) #2
```

# 5. 将3×4矩阵按行分割成3个1×4的数组。

```python
arr=np.arange(12).reshape(3,4)
print(arr)
log=np.split(arr,3,axis=0)
print(log)
for i in log:  #1
    print(i)
l=np.vsplit(arr,3)
for i in l:  #2
    print(i)
```

# 6. 将3×4矩阵按列分割成2个3×2的数组。

```python
arr=np.arange(12).reshape(3,4)
print(arr)
print(np.split(arr,2,axis=1))  #1  
print(np.hsplit(arr,2))# 2
```

# 7. 使用`newaxis`将一维数组(5,)变成列向量(5,1)。

```python
a=np.random.randint(0,100,5)
print(a)
print(a[:,np.newaxis])
print(a[np.newaxis,:])
```

# 8. 使用`concatenate`沿指定轴合并两个三维数组。

```python
arr1=np.random.randn(2,3,4)
arr2=np.random.randn(2,3,4)
print(np.concatenate((arr1,arr2),axis=0))
```

# 9. 给数组增加一个维度（例如从(3,4)变成(1,3,4)）

```python
np.random.seed(520)
arr=np.random.randint(0,100,(3,4))
print(arr)
a=arr[np.newaxis,:,:]
print(a)
print(a.shape)
```

# 10. 删除数组中大小为1的维度（`squeeze`）。

```python
print(a.shape)
b=np.squeeze(a)
print(b)
print(b.shape)
```
