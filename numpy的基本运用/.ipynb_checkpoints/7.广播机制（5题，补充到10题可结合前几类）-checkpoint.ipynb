```python
import numpy as np
```

# 1. 创建一个3×1的数组和一个1×3的数组，相加后观察广播结果。

```python
np.random.seed(520)
arr1=np.random.randint(0,10,(3,1))
arr2=np.random.randint(0,10,(1,3))
print(arr1)
print(arr2)
print(arr1+arr2)
```

# 2. 对一个3×4矩阵，加上一个长度为4的一维向量，解释广播规则。

```python
np.random.seed(521)
arr1=np.random.randint(0,10,(3,4))
arr2=np.random.randint(0,10,4)
print(arr1)
print(arr2)
print(arr1+arr2)   #如果两个数组的维度不同，从尾部开始比较，若相等或一方为1则可广播
```

# 3. 对一个3×4矩阵，加上一个长度为3的一维向量，能否广播？如何调整？

```python
np.random.seed(520)
arr1=np.random.randint(0,10,(3,4))
arr2=np.random.randint(0,10,3)
print(arr1)
print(arr2)
try:
    arr1+arr2
except ValueError as e:
    print("错误",e) 
print(arr1+arr2.reshape(3,1))
```

# 4. 使用广播将矩阵的每一行减去该行的均值（中心化）。

```python
np.random.seed(520)
arr=np.random.randint(0,10,(3,4))
print(arr)
a=np.mean(arr,axis=1,keepdims=True)
print(a)
print(arr-a)
```

# 5. 使用广播将矩阵的每一列除以该列的标准差（标准化）。

```python
a=np.std(arr,axis=0,keepdims=True)
print(a)
print(arr/a)
```

# 6. 计算两个形状分别为(3,1)和(1,4)的数组的外积（用广播）。

```python
np.random.seed(520)
a=np.random.randint(0,10,(3,1))
b=np.random.randint(0,10,(1,4))
print(a)
print(b)
print(np.outer(a,b))
```

# 7. 创建一个3×4矩阵，将其每一元素乘以一个标量（自然广播）。

```python
np.random.seed(520)
arr=np.random.randint(1,5,(3,4))
print(arr)
print(arr*10)
```

# 8. 创建一个3×4矩阵和一个3×1数组，相除。

```python
np.random.seed(520)
arr1=np.random.randint(1,10,(3,4))
arr2=np.random.randint(1,10,(3,1))
print(arr1)
print(arr2)
print(arr1/arr2)
```

# 9. 创建一个4×3矩阵和一个3×2矩阵，利用广播实现某种操作（需设计）。

```python
np.random.seed(520)
arr=np.random.randint(1,10,(4,3))
arr2=np.random.randint(1,10,(3,2))
print(arr)
print(arr2)
print(arr@arr2)
```

```python
b=arr[:,:,np.newaxis]+arr2[np.newaxis,:,:]
print(b)
print(b.shape)
```
