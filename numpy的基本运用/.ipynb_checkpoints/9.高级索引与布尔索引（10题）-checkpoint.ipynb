```python
import numpy as np
```

# 1. 使用布尔数组提取数组中所有大于0的元素。

```python
np.random.seed(520)
arr=np.random.randint(0,3,12)
print(arr)
a=arr>0
print(a)
print(arr[a])
```

# 2. 将数组中所有非数字（NaN）替换为0。

```python
arr=np.array([np.nan,12,545,745,np.nan,55,np.nan])
print(np.isnan(arr))
arr[np.isnan(arr)]=0
print(arr)
```

# 3. 统计数组中大于某个阈值的元素个数。

```python
np.random.seed(520)
arr=np.random.randint(0,50,20)
print(arr)
count=np.sum(arr>=25)
print(count)
print(np.sum(np.where(arr>=25,1,0)))
```

# 4. 使用布尔索引修改满足条件的元素（如将所有奇数变为-1）。

```python
np.random.seed(520)
arr=np.random.randint(0,50,20)
print(arr)
print(np.where(arr%2!=0,-1,arr))
```

# 5. 找出数组中同时满足两个条件的元素（例如大于5且小于10）。

```python
np.random.seed(520)
arr=np.random.randint(0,20,20)
print(arr)
a=(arr>=5)&(arr<=10)
print(arr[a])
```

# 6. 用花式索引提取指定位置的元素（索引数组）。

```python
np.random.seed(520)
arr=np.random.randint(0,20,20)
print(arr)
print(arr[[2,5,6]])
```

# 7. 创建一个3×3矩阵，用两个一维索引数组（行索引和列索引）提取对应元素。

```python
np.random.seed(520)
arr=np.random.randint(0,20,(3,3))
print(arr)
a1=[0,1,2]
a2=[0,1,2]
print(arr[a1,a2])
```

# 8. 使用`take`和`put`函数进行索引和赋值。

```python
a = np.array([10,20,30,40])
indices = [0,2]
taken = a.take(indices)
print(taken)
a.put([1,3], [99,88])
print(a)
```

# 9. 使用`ix_`函数构造索引器，提取不连续的行和列。

```python
np.random.seed(520)
arr=np.random.randint(0,3,(5,5))
print(arr)
rows=[1,2,3]
cols=[0,4]
print(np.ix_(rows,cols))
print(arr[1,0],arr[1,4],arr[2,0],arr[2,4],arr[3,0],arr[3,4])
print(arr[np.ix_(rows,cols)])
```

# 10. 对一维数组，找出局部最大值的索引（即比前后元素都大的位置）

```python
np.random.seed(520)
arr=np.random.randint(0,100,10)
print(arr)
js=[]
ji=[]
for i in range(1,len(arr)-1):
    if arr[i]>arr[i-1] and arr[i]>arr[i+1]:
        ji.append(i)
        js.append(arr[i])
print(js)
a=(arr[1:-1]>arr[:-2])&(arr[1:-1]>arr[2:])
print(a)
p = np.arange(1, len(arr)-1)[a]
print(ji) 
print(p)
```
