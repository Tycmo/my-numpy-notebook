# 1. 计算两个2×2矩阵的乘积（`dot`或`@`）

```python
np.random.seed(520)
arr=np.random.randint(0,100,(2,2))
arr2=np.random.randint(0,100,(2,2))
print(arr)
print(arr2)
print(arr@arr2)
print(np.dot(arr,arr2))
```

# 2. 计算3×3矩阵的行列式。

```python
arr=np.random.randint(0,10,(3,3))
print(arr)
print(np.linalg.det(arr))
```

# 3. 求3×3矩阵的逆矩阵。

```python
print(np.linalg.inv(arr))
```

# 4. 解线性方程组：  
   \[
   \begin{cases}
   2x + y = 5 \\
   x - 3y = 6
   \end{cases}
   \]  
   用矩阵形式`Ax = b`求解。

```python
A=np.array([[2,1],[1,-3]])
b=np.array([5,6])
x=b@np.linalg.inv(A)
print(x)
```

# 5. 计算矩阵的特征值和特征向量。

```python
arr=np.arange(16).reshape(4,4)
print(arr)
a,b=np.linalg.eig(arr)
print(a,b)
```

# 6. 计算两个向量的内积和外积。

```python
np.random.seed(520)
a=np.random.randint(0,5,(3,4))
b=np.random.randint(0,5,(3,4))  
print(a)
print(b)            # 内积：逐元素相乘后的和，结果为一个标量
print(np.sum(a*b))  # 对应多维
a=np.array([1,2,3]) # 对应一维
b=np.array([4,5,6])
print(np.dot(a, b))
print(a @ b)
print(np.inner(a, b))    
print("\n", "外积")  #
print(np.outer(a,b))
```

# 7. 对矩阵进行奇异值分解（SVD）

```python
a=np.array([[1,2],[3,4],[5,6]])
u,s,vt=np.linalg.svd(a)
print(a)
print(u)
print(s)
print(vt)
```

# 8. 计算矩阵的迹（对角线元素之和）。

```python
arr=np.arange(12).reshape(3,4)
print(arr)
print(np.trace(arr))
```

# 9. 求矩阵的范数（Frobenius范数）。

```python
arr=np.arange(12).reshape(3,4)
print(arr)
print(np.linalg.norm(arr))
```
