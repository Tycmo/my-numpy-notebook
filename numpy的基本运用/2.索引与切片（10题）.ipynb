```python
import numpy as np
```

# 1.创建一个5×5的随机矩阵，提取第2行（索引从0开始

```python
np.random.seed(520)
arr=np.random.rand(5,5)
print(arr)
print("\n")
print(arr[1])  #1
print(arr[1,:])  #2
```

# 2. 提取上述矩阵的第1列和第3列。

```python
print(arr[:,0])
print(arr[:,2])
```

# 3. 提取矩阵的2-4行、1-3列（左闭右开）。

```python
print(arr[1:4,:]) #连续取法
print(arr[:,0:3])
print(arr[[1,3],:]) #非连续取法
print(arr[:,[0,2]])
```

# 4. 提取矩阵中所有偶数行、奇数列。

```python
print("偶数")
print(arr[1::2,:])   #步长
```

# 5. 将矩阵的中心3×3子矩阵全部赋值为0。

```python
arr[1:4,1:4]=0
print(arr)
```

# 6. 使用花式索引提取指定行[0,2,4]。

```python
print(arr[[0,2,4],:])
```

# 7. 使用花式索引提取指定列[1,3]并组成新矩阵。

```python
ar=arr[:,[1,3]]
print(ar)
```

# 8. 创建一个一维数组，提取所有大于10的元素。

```python
arr=np.random.randint(-21,21,20)
print(arr)
```

```python
print(arr[arr>10])  #1
print(np.extract(arr>10,arr))  #2
```

# 9. 将数组中所有负数替换为0。

```python
arr3=np.where(arr<0,0,arr)  #1
print(arr3)
arr[arr<0]=0 #2
print(arr)
```

# 10. 获取数组中最大值和最小值的索引（使用`argmax`和`argmin`）。

```python
print(np.argmin(arr))
print(np.argmax(arr))
```
